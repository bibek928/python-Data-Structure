1. What are data structures, and why are they important?
    
    - Data structures are ways of organizing and storing data so it can be accessed and modified efficiently. They are important because they optimize data handling for operations like searching, sorting, insertion, and deletion.
2. Explain the difference between mutable and immutable data types with examples.
  - Mutable: Can be changed (e.g., List → a = [1, 2]; a[0] = 10)

  Immutable: Cannot be changed (e.g., String, Tuple → a = "hello"; a[0] = 'H' will cause an error)
3. Differences between lists and tuples in Python?

   - Lists are mutable; tuples are immutable.
   - Lists have more built-in methods.
   Tuples are generally faster and used for fixed data.
4. How dictionaries store data?
  - Dictionaries store data as key-value pairs. Internally, they use a hash table where each key is hashed to determine where its value is stored.

5. Why use a set instead of a list in Python?
   - Sets are used to store unique elements and perform set operations (like union, intersection) efficiently.

6. What is a string in Python, and how is it different from a list?
   - A string is a sequence of characters and is immutable, while a list is mutable and can contain mixed data types.

7. How do tuples ensure data integrity in Python?
   - Their immutability prevents accidental changes, making them ideal for constants or fixed collections.

8. What is a hash table, and how does it relate to dictionaries in Python?
   - A hash table is a data structure that maps keys to values using a hash function. Python dictionaries are implemented as hash tables.

9. Can lists contain different data types in Python?
   - Yes, lists in Python can contain elements of different data types.

10. Why are strings immutable in Python?
  - To improve performance and security. Since strings are often reused, immutability allows them to be safely shared.

11. Advantages of dictionaries over lists for certain tasks?
  - Fast key-based access (O(1) on average) vs. O(n) lookup in lists.

12. Scenario where using a tuple is preferable over a list?
  - When storing coordinates (x, y), or passing fixed arguments that shouldn’t change.

13. How do sets handle duplicate values?
  - Sets automatically eliminate duplicates.

14. How does the “in” keyword work differently for lists and dictionaries?
  - List: checks for value presence.
  - Dict: checks for key presence.

15. Can you modify elements of a tuple? Why or why not?
  - No.Tuples are immutable; their contents can't be changed after creation.

16. What is a nested dictionary? Example
  - A dictionary within another dictionary.
data = {'person': {'name': 'Alice', 'age': 30}}

17. Time complexity of accessing elements in a dictionary?
- O(1) on average due to hashing.

18. When are lists preferred over dictionaries?
 - When order matters or when storing simple sequences without key-value mapping.

19. Why are dictionaries unordered and how it affects retrieval?
 - Before Python 3.7, they were unordered. Now, they preserve insertion
 order, but traditional retrieval was based on key, not index.

20. Difference between a list and a dictionary in data retrieval?
 - Lists use index; dictionaries use key.
